```python
#imports
import numpy as np
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt 
from sklearn.impute import SimpleImputer
import statsmodels.api as sm
from sklearn.preprocessing import LabelEncoder 
import random

le = LabelEncoder()
```

```python

modeldata = pd.read_csv("C:\Programming Assignment\mergedDF.csv")
pd.set_option('display.max_rows', 195)


##QA test indicated that null values are ruining the prediction model therefore i must slice the data frame until name and bikestations are consistent 
sliced_modeldata = modeldata.iloc[:, :145]
sliced_modeldata
```

```python


sliced_modeldata['rating'].fillna(random.uniform(2.5, 4.8), inplace=True)
sliced_modeldata['free_bikes'].fillna(random.uniform(0, 12), inplace=True)
```

```python
sliced_modeldata['rating'] = sliced_modeldata['rating'].round(1)
sliced_modeldata['free_bikes'] = sliced_modeldata['free_bikes'].round(0)
```

```python
sliced_modeldata
```

```python
sliced_modeldata = sliced_modeldata.sort_values(by='rating', ascending=False)
sliced_modeldata.reset_index(drop=True, inplace=True)
```

```python
xdata = sliced_modeldata[['rating']]
ydata = sliced_modeldata['free_bikes']
df_xdata = pd.DataFrame(xdata)
df_ydata = pd.Series(ydata)
```
